```python
import numpy as np
import pandas as pd
#construct a BoW method based on word count
from sklearn.feature_extraction.text import CountVectorizer
#this method allows us to determine significant words in a document after Countvectorizer
#term frequency - inverse document frequency.
#This score indicates the importance of a term within a specific document compared to its occurrence across all documents in the corpus
from sklearn.feature_extraction.text import TfidfTransformer
#this library allows us to remove unwanted chars from a text
# regular expression or regx
import re
#this library allows us to stemm (transform words into its base or root)
from nltk.stem.porter import PorterStemmer
#Stop word method allows us to remove words like e is, and, has, and like , the word which bear no useful information
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('/content/ACM.csv')
```

```python
data.head(3)
```

```python
# Bag-of-words which represents text as numerical form
```

```python
# before creating BoW it is vital to clear the text by removing unwanted character
```

```python
#to accomplish this task we use pythons regular expresion
```

```python
# normalization , remove unwanted charachters
def preproccesor(text):
  text = re.sub('<[^>]*>', '', text)
  emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)',text)
  text = (re.sub('[\W]+', ' ', text.lower()) +' '.join(emoticons).replace('-', ''))
  return text
```

```python
data['title'] = data['title'].apply(preproccesor)
```

```python
# now it is time to tokenize the text
```

```python
def tokenizer(text):
  return text.split()
```

```python
data['title'] = data['title'].apply(tokenizer)
```

```python
data.sample(10)
```

```python
# in the context of tokenization
#an other useful technique is word stemming
# which is transforming word to its base or root without any suffix and prefix
```

```python
porter = PorterStemmer()
```

```python
def tokenizer_porter(text):
    porter = PorterStemmer()
    return [porter.stem(word) for word in str(text).split()]  # Convert to string before splitting
```

```python
data['title'] = data['title'].apply(tokenizer_porter)
```

```python
stop = stopwords.words('english')
data['title'] = [w for w in data['title']if w not in stop]
```

```python
data.sample(5)
```

```python
vectorizer = TfidfVectorizer(strip_accents=None,lowercase=False,preprocessor=None)
```

```python
data['title'] = data['title'].astype(str)
tfidf_matrix = vectorizer.fit_transform(data['title'])
```
